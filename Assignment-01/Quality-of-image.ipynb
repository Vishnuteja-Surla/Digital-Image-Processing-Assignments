# Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Constants

```python
max_pixel = 255.0
```

# Reading the image

```python
original_image = cv2.imread("Images/my-face.png")
```

```python
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()
```

# Degrading the image

### 1. Blurring the image

```python
ksize = (5, 5)
blurred_image = cv2.blur(original_image, ksize)
```

```python
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title("Blurred Image")
plt.show()
```

### 2. Adding Gaussian Noise

```python
# Generate random Gaussian noise
mean = 0
stddev = 180
noise = np.zeros(blurred_image.shape, np.uint8)
cv2.randn(noise, mean, stddev)
```

```python
noisy_image = cv2.add(blurred_image, noise)
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.title("Noisy Image")
plt.show()
```

### 3. Compression to jpeg and decompressing around 70-80% quality

```python
cv2.imwrite('Images/degraded-image.jp2', noisy_image, [int(cv2.IMWRITE_JPEG2000_COMPRESSION_X1000), 70])
```

```python
degraded_image = cv2.imread("Images/degraded-image.jp2")
```

# Image Quality difference identification

```python
mse = np.mean((original_image - degraded_image)**2)
print(f"MSE = {mse}")
```

```python
psnr = 20 * np.log10(max_pixel/np.sqrt(mse))
print(f"PSNR = {psnr} dB")
```
